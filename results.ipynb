```python
def plot_row(dataframes,names,title):
    """
    This function takes a list of Pandas DataFrames with one row each 
    and plots the values in this row.
    
    Parameters:
        dataframes (list): List of Pandas DataFrames, each containing only one row.
    
    Returns:
        None
    """
    fig, ax = plt.subplots()
    
    for i, df in enumerate(dataframes):
        # Plot the values
        ax.plot(df.columns, df.iloc[0], marker='o', label=names[i])
    
    ax.set_xlabel('Epsilon')
    ax.set_ylabel('Accuracy')
    plt.yticks(np.arange(0.0, 1.1, 0.1))
    ax.legend()
    plt.title(title)
    ax.grid(True)
    f_name = title.replace(' ','_')
    plt.savefig(f'{f_name}.pdf',format='pdf')
    plt.show()
    


# plot_row([results_base,results_HiFiC,results_HiFiC_through,results_jpeg,results_jpeg_through],['Baseline','HiFiC','HiFiC_attacked','JPEG','JPEG_attacked'],'iFGSM attack')
```

```python
import os
dfs = []
names = []
path = 'results/jpeg_seq/'
for f in os.listdir(path):
    dfs.append(pd.read_csv(path+f,index_col = 0))
    names.append(f.split('_')[2])
```

```python

plot_row(dfs,names,'iFGSM sequential jpeg with gradient information')
```

```python
eps8 = [df.iloc[0,2] for df in dfs]
baseline = [df.iloc[0,0] for df in dfs]
eps4 = [df.iloc[0,1] for df in dfs]
eps16 = [df.iloc[0,3] for df in dfs]
```

```python
names_i = [int(n) for n in names]
```

```python
s_names= [names[i] for i in np.argsort(names_i)]
s_baseline = [baseline[i] for i in np.argsort(names_i)]
s_eps4 = [eps4[i] for i in np.argsort(names_i)]
s_eps8 = [eps8[i] for i in np.argsort(names_i)]
s_eps16 = [eps16[i] for i in np.argsort(names_i)]
```

```python
fig, ax = plt.subplots()
ax.plot(s_names,s_baseline, marker='o', label='baseline')
ax.plot(s_names,s_eps4, marker='o', label='4/255')
ax.plot(s_names,s_eps8, marker='o', label='8/255')
ax.plot(s_names,s_eps16, marker='o', label='16/255')
ax.set_xlabel('Times thourgh JPEG')
ax.set_ylabel('Accuracy')
ax.legend()
plt.yticks(np.arange(0.0, 1.1, 0.1))
plt.title('Accuracy for iFGSM for different N of JPEG compression in sequence')
ax.grid(True)
plt.savefig('sequential_jpeg.pdf',format='pdf')
```
